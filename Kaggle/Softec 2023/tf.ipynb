```python
import pandas as pd
import tensorflow as tf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
description = pd.read_excel('Softec_Attribute_Description.xlsx')
dfA=pd.read_csv('Softec_Medication_Data_Kaggle_V1.csv')
dfB=pd.read_csv('Softec_Patient_Data_Testing_Features.csv')
Ans=pd.read_csv('Softec_Patient_Data_Testing_Kaggle_V1.csv')
dfD=pd.read_csv('Softec_Patient_Data_Training_Kaggle_V1.csv')
dfE=pd.read_csv('Softec_Socio_Economics_Data_Kaggle_V1.csv')
```

```python
Test=pd.merge(pd.merge(dfA,dfE,on='ID'),dfB,on='ID')
Train=pd.merge(pd.merge(dfA,dfE,on='ID'),dfD,on='ID')
```

```python
Train.Race=Train.Race.replace({'Black': 0, 'White': 1,'Hispanic':2,'Asian':3,'Other Race':4,'North American Native':5,'UnKnown':11})
Train.Race.astype(int)
Train.Race=Train.Race.replace({11:np.nan})
Train.Sex=Train.Sex.replace({'Male':0,'Female':1})
Train.Sex.astype(int)
Train.State=pd.factorize(Train['State'])[0]
Train.MSA=pd.factorize(Train['MSA'])[0] #8th value is unkown 148 (values are unknown if not too many either drop those or fill those out before using corr)
Train.MSA=Train.MSA.replace({8:np.nan})
Train['Enrollment Type Categorized']=pd.factorize(Train['Enrollment Type Categorized'])[0]
Train=Train.drop('Avg ADI',axis=1)
Train=Train.drop('Max ADI',axis=1)
Train=Train.drop('Min ADI',axis=1)
Train=Train.drop('SVI1',axis=1)
Train=Train.drop('SVI3',axis=1)
Train=Train.drop('SVI4',axis=1)
Train=Train.drop('SVI',axis=1)
Train=Train.drop('MSA',axis=1)
Train=Train.drop('MDI',axis=1)
Train=Train.drop('Race',axis=1)
Train=Train.drop('State',axis=1)
Train=Train.drop('Zip Code',axis=1)
```

```python
bins = [-float('inf'), 1, 2, 3, 4, 7, 14, float('inf')]
labels = [0, 1, 2, 3, 4, 5, 7]
```

```python
Train['# Rehabilitation Hospital'].value_counts()
```

```python
Train['HCC Score'] = np.where(Train['HCC Score'] < 1, 0, 1)
Train['Age'] = np.where(Train['Age'] < 64, 0, 1)
Train['TotalCost_Y_Actual'] = np.where(Train['TotalCost_Y_Actual'] < 11000, 0, 1)
Train['TotalCost_Y_Expected'] = np.where(Train['TotalCost_Y_Expected'] < 11000, 0, 1)
Train['Diagnosis'] = pd.cut(Train['Diagnosis'], bins=bins, labels=labels, right=False)
Train['Diagnosis']=Train['Diagnosis'].astype(int)
Train['# Home Health'] = pd.cut(Train['# Home Health'], bins=bins, labels=labels, right=False)
Train['# Home Health']=Train['# Home Health'].astype(int)
Train['# Labs'] = pd.cut(Train['# Labs'], bins=bins, labels=labels, right=False)
Train['# Labs']=Train['# Labs'].astype(int)
Train['Enrollment Months'] = pd.cut(Train['Enrollment Months'], bins=bins, labels=labels, right=False)
Train['Enrollment Months']=Train['Enrollment Months'].astype(int)
Train['Avg. LOS'] = pd.cut(Train['Avg. LOS'], bins=bins, labels=labels, right=False)
Train['Avg. LOS']=Train['Avg. LOS'].astype(int)
Train['# Total Claims'] = pd.cut(Train['# Total Claims'], bins=bins, labels=labels, right=False)
Train['# Total Claims']=Train['# Total Claims'].astype(int)
Train['# Hospital OP'] = pd.cut(Train['# Hospital OP'], bins=bins, labels=labels, right=False)
Train['# Hospital OP']=Train['# Hospital OP'].astype(int)
Train['# All Physician OP'] = pd.cut(Train['# All Physician OP'], bins=bins, labels=labels, right=False)
Train['# All Physician OP']=Train['# All Physician OP'].astype(int)
Train['# Inpatient'] = pd.cut(Train['# Inpatient'], bins=bins, labels=labels, right=False)
#Train['# Inpatient']=Train['# Inpatient'].astype(int)
```
