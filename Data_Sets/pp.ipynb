```python
import numpy as np
import pandas as pd
import random
```

```python
df=pd.read_csv('Dif_Format_dataset.csv')
```

```python
p=[]
c=[]
for i in range(len(df)):
    p.append(df.loc[i].Prompt)
    c.append(df.loc[i].Completion)
```

```python
n=random.randint(0,len(df))
p[n]+c[n]
```

```python
f=open('extraNonhistPrompts.txt','r')
s=f.readline()
flag=False
while s!='':
    if(s=='\n'):
        s=f.readline()
        continue
    s=s.rstrip('\n')
    if flag==False:
        flag=True
        p.append(s)
    else:
        flag=False
        c.append(s)
    s=f.readline()
```
